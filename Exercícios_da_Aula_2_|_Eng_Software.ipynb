# PARTE 2: Implementação do Sistema Conversor de Temperatura

```python
# ====================================
# CONVERSOR DE TEMPERATURA
# Exercício - Engenharia de Software
# ====================================
def celsius_para_fahrenheit(celsius):
  fahrenheit = (celsius * 9/5) + 32
  return fahrenheit

def fahrenheit_para_celsius(fahrenheit):
  celsius = (fahrenheit - 32) * 5/9
  return celsius

def kelvin_para_celsius(kelvin):
  celsius = kelvin - 273.15
  return celsius

def celsius_para_kelvin(celsius):
  kelvin = celsius + 273.15
  return kelvin

def kelvin_para_fahrenheit(kelvin):
  fahrenheit = kelvin * 1.8 - 459.67
  return fahrenheit

def fahrenheit_para_kelvin(fahrenheit):
  kelvin = (fahrenheit + 459.67) / 1.8
  return kelvin

decisao = 1

while decisao == 1:
  print("\nOpção 1: Conversão de Celsius → Fahrenheit")
  print("Opção 2: Conversão de Fahrenheit → Celsius")
  print("Opção 3: Conversão de Kelvin → Celsius")
  print("Opção 4: Conversão de Celsius → Kelvin")
  print("Opção 5: Conversão de Kelvin → Fahrenheit")
  print("Opção 6: Conversão de Fahrenheit → Kelvin")
  print("\n")
  opcao = int(input("Digite a opção desejada: "))
  print("\n")

  if opcao == 1:
    celsius = float(input("Digite a temperatura em Celsius: "))
    calculo = celsius_para_fahrenheit(celsius)
    print("\n")
    print("="*40)
    print(f"{celsius}°C = {calculo:.2f}°F")
    print("="*40)

  elif opcao == 2:
    fahrenheit = float(input("Digite a temperatura em Fahrenheit: "))
    calculo = fahrenheit_para_celsius(fahrenheit)
    print("\n")
    print("="*40)
    print(f"{fahrenheit}°F = {calculo:.2f}°C")
    print("="*40)

  elif opcao == 3:
    kelvin = float(input("Digite a temperatura em Kelvin: "))
    calculo = kelvin_para_celsius(kelvin)
    print("\n")
    print("="*40)
    print(f"{kelvin} K = {calculo:.2f}°C")
    print("="*40)

  elif opcao == 4:
    celsius = float(input("Digite a temperatura em Celsius: "))
    calculo = celsius_para_kelvin(celsius)
    print("\n")
    print("="*40)
    print(f"{celsius}°C = {calculo:.2f} K")
    print("="*40)

  elif opcao == 5:
    kelvin = float(input("Digite a temperatura em Kelvin: "))
    calculo = kelvin_para_fahrenheit(kelvin)
    print("\n")
    print("="*40)
    print(f"{kelvin} K = {calculo:.2f}°F")
    print("="*40)

  elif opcao == 6:
    fahrenheit = float(input("Digite a temperatura em Fahrenheit: "))
    calculo = fahrenheit_para_kelvin(fahrenheit)
    print("\n")
    print("="*40)
    print(f"{fahrenheit}°F = {calculo:.2f} K")
    print("="*40)

  else:
    print("\nOpção inválida")

  print("\nDeseja continuar realizando conversões?: (1-sim) (0-não)")
  decisao = int(input())

if decisao == 0:
  print("\nObrigado por utilizar nosso sistema!")
  print("Finalizando o sistema ......")
```

**Comentário sobre o output exibido:**
<br>
O output demonstra que o sistema executou corretamente a conversão de 23°C para 73,40°F, exibindo o resultado formatado com duas casas decimais. Além disso, o programa apresentou uma navegação clara e permitiu ao usuário encerrar o sistema de forma simples.
