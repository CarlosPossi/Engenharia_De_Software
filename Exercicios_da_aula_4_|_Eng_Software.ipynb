# 🎯 Exercício Prático — Hands-on!

###🚀 Missão: Construir o SRS do "FIAP Marketplace"
Contexto: A FIAP quer criar um marketplace interno onde alunos podem vender produtos artesanais entre si (como um Etsy da FIAP 🛒).
<br>
<br>

#### **Parte 1 — Estender o código Python da aula**

```python
from dataclasses import dataclass, field
from typing import List
from enum import Enum

class Prioridade(Enum):
    ALTA = "Alta"
    MEDIA = "Média"
    BAIXA = "Baixa"

@dataclass
class RequisitoFuncional:
    id: str
    nome: str
    descricao: str
    prioridade: Prioridade
    ator: str
    pre_condicao: str
    pos_condicao: str

@dataclass
class RequisitoNaoFuncional:
    id: str
    categoria: str  # Desempenho, Segurança, Usabilidade...
    descricao: str
    criterio_aceitacao: str

@dataclass
class SRS:
    projeto: str
    versao: str
    descricao: str
    requisitos_funcionais: List[RequisitoFuncional] = field(default_factory=list)
    requisitos_nao_funcionais: List[RequisitoNaoFuncional] = field(default_factory=list)

    def adicionar_rf(self, req: RequisitoFuncional):
        self.requisitos_funcionais.append(req)
        print(f"✅ '{req.id}' adicionado!")

    def adicionar_rnf(self, req: RequisitoNaoFuncional):
        self.requisitos_nao_funcionais.append(req)
        print(f"✅ '{req.id}' adicionado!")

    def relatorio(self):
        print(f"\n{'='*145}")
        print(f"🛒 SRS — {self.projeto} V {self.versao}")
        print(f"{'='*145}")
        print(f"{self.descricao}\n")

        print(f"\n🔧 REQUISITOS FUNCIONAIS\n")
        for rf in self.requisitos_funcionais:
            print(f"  [{rf.id}] {rf.nome} — Prioridade: {rf.prioridade.value}")
            print(f"       👤 Ator: {rf.ator}")
            print(f"       📌 {rf.descricao}\n")

        print(f"\n\n⚡ REQUISITOS NÃO-FUNCIONAIS\n")
        for rnf in self.requisitos_nao_funcionais:
            print(f"  [{rnf.id}] - {rnf.categoria}")
            print(f"       📌 {rnf.descricao}")
            print(f"       ✔️  Critério: {rnf.criterio_aceitacao}\n")

# ---- Criando o SRS do App FIAP Marketplace ---

srs = SRS(
    projeto="App FIAP Marketplace — Módulo de Vendas de Produtos",
    versao="1.0",
    descricao="Sistema de marketplace interno onde alunos/professores podem realizar vendas de produtos"
)

srs.adicionar_rf(RequisitoFuncional(
    id="RF-001",
    nome="Cadastramento de Vendedores",
    descricao="O vendedor deverá realizar o seu cadastramento, com suas informações pessoais",
    prioridade=Prioridade.MEDIA,
    ator="Vendedor",
    pre_condicao="O vendedor deverá ter um ou mais produtos cadastrados em seu catálogo",
    pos_condicao="O vendedor poderá realizar a venda para qualquer usuário"
))

srs.adicionar_rf(RequisitoFuncional(
    id="RF-002",
    nome="Cadastramento de Consumidores",
    descricao="O consumidor deverá realizar o cadastramento, com suas informações pessoais",
    prioridade=Prioridade.BAIXA,
    ator="Consumidor",
    pre_condicao="As informações pessoais devem estar todas preenchidas",
    pos_condicao="Cliente está pronto para realizar qualquer compra"
))

srs.adicionar_rf(RequisitoFuncional(
    id="RF-003",
    nome="Cadastramento de Produtos",
    descricao="O vendedor deverá realizar o cadastramento de seus produtos",
    prioridade=Prioridade.ALTA,
    ator="Vendedor",
    pre_condicao="O número de produtos para venda, deve ser maior que uma unidade",
    pos_condicao="Cliente pode realizar a compra"
))

srs.adicionar_rf(RequisitoFuncional(
    id="RF-004",
    nome="Notificação de Confirmação/Recusa de compra",
    descricao="Enviar push notification ao cliente quando status da compra for alterado.",
    prioridade=Prioridade.ALTA,
    ator="Sistema",
    pre_condicao="Cliente com notificações habilitadas",
    pos_condicao="Cliente notificado sobre status da compra"
))

srs.adicionar_rf(RequisitoFuncional(
    id="RF-005",
    nome="Busca por categoria",
    descricao="Usuário deve conseguir navegar pelas categorias",
    prioridade=Prioridade.MEDIA,
    ator="Usuário",
    pre_condicao="Cliente cadastrado na FIAP Marketplace",
    pos_condicao="Cliente poderá navegar pelas categorias presentes"
))

srs.adicionar_rnf(RequisitoNaoFuncional(
    id="RNF-001",
    categoria="Desempenho",
    descricao="O sistema deve suportar 30.000 usuários simultâneos.",
    criterio_aceitacao="Teste de carga com JMeter: 30k req/s com latência < 800ms"
))

srs.adicionar_rnf(RequisitoNaoFuncional(
    id="RNF-002",
    categoria="Segurança",
    descricao="Dados pessoais dos clientes/vendedores devem ser criptografados em trânsito.",
    criterio_aceitacao="Uso de TLS 1.3 validado por ferramenta de auditoria"
))

srs.adicionar_rnf(RequisitoNaoFuncional(
    id="RNF-003",
    categoria="Usabilidade",
    descricao="A interface deve ser simples e de fácil entendimento, permitindo que os usuários concluam a compra em menos de 15 segundos.",
    criterio_aceitacao="Usuários conseguem finalizar a compra em até 15 segundos em testes de usabilidade com pelo menos 90% de sucesso."
))

srs.adicionar_rnf(RequisitoNaoFuncional(
    id="RNF-004",
    categoria="Disponibilidade",
    descricao="O sistema de compras deve estar disponível para acesso dos usuários de forma contínua.",
    criterio_aceitacao="Disponibilidade mínima de 99,9% mensal, medida por ferramenta de monitoramento e considerando janelas de manutenção previamente programadas."
))

srs.relatorio()
```

**Comentário sobre o que o esquema representa e as decisões de modelagem:**
<br>
O diagrama representa um modelo de SRS (Software Requirements Specification) para um marketplace, estruturando requisitos funcionais e não funcionais em classes com atributos bem definidos.
A modelagem usa dataclasses para simplicidade e clareza, Enum para controlar prioridades e listas com default_factory para evitar estados mutáveis compartilhados.
Também separa responsabilidades (RF/RNF) e centraliza a gestão dos requisitos na classe SRS, facilitando extensibilidade e geração de relatórios.

**Comentário sobre o que o código implementa e o que foi aprendido:**
<br>
O diagrama representa um SRS (Software Requirements Specification) de um marketplace, organizando requisitos funcionais e não funcionais de forma estruturada.
A modelagem usa dataclasses para representar requisitos como entidades simples e Enum para padronizar prioridades.
A classe SRS centraliza o gerenciamento e a geração de relatório, separando responsabilidades e facilitando manutenção e evolução do sistema.

# Parte 2 — Análise crítica

```python
def validar_requisito(rf: RequisitoFuncional) -> dict:
    """
    Valida se um requisito funcional segue as boas práticas.
    Retorna um dict com os resultados da validação.
    """
    resultados = {}

    resultados["A descrição contém mais de 20 caracteres?"] = len(rf.descricao) > 20
    resultados["A pré-condição está definida? "] = rf.pre_condicao != ""
    resultados["Tem critério mensurável?"] = any(char.isdigit() for char in rf.descricao)

    # TODO: Implemente as validações aqui!
    # Dica 1: len(rf.descricao) > 20
    # Dica 2: rf.pre_condicao != ""
    # Dica 3: any(char.isdigit() for char in rf.descricao)

    return resultados
```

#**Parte 3** — ⭐ Desafio extra

Crie um método exportar_markdown(srs) que gera o SRS em formato Markdown, pronto para ser colado no Confluence ou Notion da sua empresa!

```python
def exportar_markdown(srs: SRS, arquivo="srs.md"):
    md = f"# 🛒 {srs.projeto} \n\n"
    md += f"{srs.descricao}\n\n"

    md += f"## ✅ Requisitos Funcionais\n"

    for rf in srs.requisitos_funcionais:
        md += f"\n\n### {rf.id} — {rf.nome}\n\n"
        md += f"❗ Prioridade: {rf.prioridade.value}\n\n"
        md += f"👤 Ator: {rf.ator}\n\n"
        md += f"📄 Descrição: {rf.descricao}\n\n"
        md += f"🔹 Pré-condição: {rf.pre_condicao}\n\n"
        md += f"🔸 Pós-condição: {rf.pos_condicao}\n\n\n\n"
        md += "<br>"

    md += f"\n\n## ☑️ Requisitos Não Funcionais\n\n"

    for rnf in srs.requisitos_nao_funcionais:
        md += f"\n\n### {rnf.id} — {rnf.categoria}\n"
        md += f"📄 Descrição: {rnf.descricao}\n\n"
        md += f"🟢 Critério de Aceitação: {rnf.criterio_aceitacao}\n\n"
        md += "<br>"

    with open(arquivo, "w", encoding="utf-8") as f:
        f.write(md)

    print(f"Arquivo '{arquivo}' gerado com sucesso!")

exportar_markdown(srs)
```
