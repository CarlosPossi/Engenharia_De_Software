## **💻 Etapa 2 — Google Colab / VS Code**
Implemente as classes em Python seguindo o diagrama que você criou no Miro.

```python
class Plataforma:
    def __init__(self, nome: str, pais: str):
        self.nome = nome
        self.pais = pais

    def iniciar(self):
        print(f"A plataforma {self.nome} foi iniciada no país {self.pais}")


class Avaliacao:
    def __init__(self, nota: float, comentario: str):
        self.nota = nota
        self.comentario = comentario

    def receberAvaliacao(self):
        print(f"Nota: {self.nota} | Comentário: {self.comentario}")


class Filme:
    def __init__(self, titulo: str, duracao: float, genero: str):
        self.titulo = titulo
        self.duracao = duracao
        self.genero = genero

    def assistir(self):
        print(f"Assistindo filme: {self.titulo}")


class Usuario:
    def __init__(self, nome: str, email: str, plano: str):
        self.nome = nome
        self.email = email
        self.plano = plano
        self.avaliacoes = []

    def avaliar(self, filme, avaliacao):
        print(f"\n👤 {self.nome} avaliou o filme {filme.titulo}")
        self.avaliacoes.append(avaliacao)

    def ver_avaliacoes(self):
        print(f"\n📝Avaliações de {self.nome}:")
        for av in self.avaliacoes:
            av.receberAvaliacao()


class Catalogo:
    def __init__(self, titulo: str, qtdFilmes: int):
        self.titulo = titulo
        self.qtdFilmes = qtdFilmes
        self.filmes = []

    def add_filme(self, filme):
        self.filmes.append(filme)
        self.qtdFilmes += 1

    def listar_filmes(self):
        print(f"🎬 {self.titulo}\n")
        for filme in self.filmes:
            print(f"📽️ Título: {filme.titulo}, Duração: {filme.duracao} min, Gênero: {filme.genero}")
```

## **📝 Deve funcionar assim:**

```python
netflix = Plataforma("Netflix", "EUA")
catalogo = Catalogo("Filmes em Destaque", 0)
filme1 = Filme("Oppenheimer", 180, "Drama")
filme2 = Filme("Barbie", 114, "Comédia")
catalogo.add_filme(filme1)
catalogo.add_filme(filme2)
usuario = Usuario("Ana", "ana@email.com", "Premium")
avaliacao = Avaliacao(9.5, "Incrível! Assisti duas vezes")
catalogo.listar_filmes()
usuario.avaliar(filme1, avaliacao)
usuario.ver_avaliacoes()
```
