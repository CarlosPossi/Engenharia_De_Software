# **🛠️ Exercício Prático — Figma Sprint 🚀**
**Contexto:** Vocês são a equipe de UX do sistema EPI da SPI Integração.
<br>
<br>
**Tarefa (individual ou dupla, ~30 min):**
<br>
### **Passo 1 — Escolha uma persona:**
🧑‍🏭 Operador | 👷 Supervisor | 📊 Gestor

### **Passo 2 — Wireframe no Figma (baixa fidelidade):**
Crie **2 telas conectadas** para a sua persona:

* Tela 1: Dashboard principal da persona
* Tela 2: Uma ação principal (ex: registrar EPI, ver alerta, gerar relatório)

### **Passo 3 — Conecte as telas:**
Use o modo Prototype para navegar da Tela 1 → Tela 2 com um clique de botão.
<br>
<br>

**Regras:**
<br>
❌ Sem cores (apenas cinza e branco)
<br>
❌ Sem imagens reais
<br>
✅ Tudo rotulado ("Botão de login", "Campo EPI", etc.)
<br>
✅ Baseado nos Requisitos Funcionais do seu Challenge
