Você foi contratado para desenvolver o sistema GymTrack — um app para academias acompanharem treinos de alunos.

# Parte 1 — Classificação
##### Liste 3 RFs e 3 RNFs para o GymTrack. Use a estrutura:

#### **Requisitos Funcionais**
* RF01: O sistema deve válidar se o nome do exercício não é vazio.
* RF02: O sistema deve válidar se o peso se encontra entre 1 e 300 kg
* RF03: O sistema deve validar se o número de repetição está entre 1 e 50

#### **Requisitos Não Funcionais**
* RNF01: O registro do exercício deve ocorrer em menos de 200 ms
* RNF02: A válidação da quantidade de peso deve em 100ms
* RNF03: A validação da quantidade de repetições em 100ms

**Comentário sobre o esquema e as decisões de modelagem:**
<br>
O esquema representa os requisitos funcionais e não funcionais do sistema GymTrack, mostrando as validações necessárias para garantir o cadastro correto dos exercícios no aplicativo. As decisões de modelagem priorizaram integridade dos dados, rapidez no processamento das validações e facilidade de uso para os usuários da academia.

# Parte 2 — Código Python

```python
import time

exercicio = str(input("Qual é o nome do exercício a ser executado?: "))
peso = float(input("Qual é a quantidade de peso utilizada nesse exercício (1 a 300kg)?: "))
repeticoes = int(input("Qual é a quantidade de repetições utilizadas nesse exercício (1 e 50)?: "))

# RF01 - O sistema deve válidar se o nome do exercício não é vazio.
if (exercicio != ""):
    mensagem = str(f"\n✅ [RF01] Exercício válido: {exercicio}")
    print(mensagem)
else:
    mensagem = str("\n❌ Exercício inválido")
    print(mensagem)


# RF02 - O sistema deve válidar se o peso se encontra entre 1 e 300 kg

if (1 <= peso <= 300):
  mensagem1 = str(f"✅ [RF02] Peso válido: {peso:.2f} kg")
  print(mensagem1)

else:
    mensagem1 = str("❌ Peso inválido")
    print(mensagem1)

# RF03 - O sistema deve validar se o número de repetição está entre 1 e 50

if (1 <= repeticoes <= 50):
    mensagem2 = str(f"✅ [RF03] Repetições válidas: {repeticoes} reps")
    print(mensagem2)

else:
    mensagem2 = str("❌ Repetições inválidas")
    print(mensagem2)

# RNF01 - O registro do exercício deve ocorrer em menos de 200 ms

inicio = time.time()
time.sleep(0.05)
if (exercicio != "" and 1 <= peso <= 300 and 1 <= repeticoes <= 50):
  print(f"\n✅ Série registrada: {exercicio} | {peso} kg x {repeticoes} reps")
else:
  print(f"\n❌ Série não registrada")
fim = time.time()
tempo_ms = (fim - inicio) * 1000

if tempo_ms < 200:
    print(f"✅ [RNF01] Tempo de registro: {tempo_ms:.0f}ms ← dentro do limite!")
else:
    print(f"❌ [RNF01] Lento demais: {tempo_ms:.0f}ms ← limite é 200ms")

#RNF02 - A válidação da quantidade de peso deve em instântanea

inicio = time.time()
time.sleep(0.05)
print(f"\n{mensagem1}")
fim = time.time()
tempo_ms = (fim - inicio) * 1000

if tempo_ms < 100:
    print(f"✅ [RNF02] Tempo de validação: {tempo_ms:.0f}ms ← dentro do limite!")
else:
    print(f"❌ [RNF02] Lento demais: {tempo_ms:.0f}ms ← limite é 100ms")

#RNF03 - A validação da quantidade de repetições deve ser em 100ms

inicio = time.time()
time.sleep(0.05)
print(f"\n{mensagem2}")
fim = time.time()
tempo_ms = (fim - inicio) * 1000

if tempo_ms < 100:
    print(f"✅ [RNF03] Tempo de validação: {tempo_ms:.0f}ms ← dentro do limite!")
else:
    print(f"❌ [RNF03] Lento demais: {tempo_ms:.0f}ms ← limite é 100ms")

```

**Comentário sobre o que o código implementa e o que foi aprendido:**
<br>
O código implementa um sistema de registro e validação de exercícios físicos, verificando regras funcionais como nome do exercício, faixa de peso e número de repetições.
Além disso, simula requisitos não funcionais ao medir o tempo de execução de cada etapa, usando time para avaliar desempenho do sistema.
O principal aprendizado é a diferença entre requisitos funcionais e não funcionais na prática, e como métricas de tempo podem ser usadas para validar desempenho além da lógica de negócio.

**Comentário sobre o output exibido:**
<br>
O output demonstra que o sistema validou corretamente o nome do exercício, o peso informado e a quantidade de repetições conforme os requisitos funcionais definidos. Além disso, os tempos de processamento ficaram dentro dos limites estabelecidos pelos requisitos não funcionais, garantindo eficiência e rapidez no registro da série.
